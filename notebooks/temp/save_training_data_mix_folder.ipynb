```python
path_table_input = '/data/kanferg/Images/Pex_project/Basyein_classifcation/output_lab_meeting/Kernel_80_mix_0to6'
path_txt = '/data/kanferg/Images/Pex_project/Basyein_classifcation/output_lab_meeting/Kernel_80_mix_0to6/png'
def upload_table(image_num,path,path_txt):
    table = pd.read_csv(os.path.join(path,'table'+str(image_num) +'.csv'))
    table['pheno_class'] = 0
    table['image_number'] = image_num
    with open(os.path.join(path_txt,'class_'+str(image_num)+'.txt')) as f:
        text = f.readline()
    tt = [int(tx) for tx in text.split(';')]
    table.iloc[tt,5] = 1
    return table
table0 = upload_table(image_num = 0,path = path_table_input,path_txt = path_txt)    
table1 = upload_table(image_num = 1,path = path_table_input,path_txt = path_txt)  
table2 = upload_table(image_num = 2,path = path_table_input,path_txt = path_txt)
table3 = upload_table(image_num = 3,path = path_table_input,path_txt = path_txt)
table4 = upload_table(image_num = 4,path = path_table_input,path_txt = path_txt)
table5 = upload_table(image_num = 5,path = path_table_input,path_txt = path_txt)
table6 = upload_table(image_num = 6,path = path_table_input,path_txt = path_txt)
```

```python
# upload image_table of the granularity mesure
def upload_img_table(image_num,path):
    table = pd.read_csv(os.path.join(path,'table_img'+str(image_num) +'.csv'))
    table['fov_number'] = image_num
    return table
table_img0 = upload_img_table(image_num = 0, path = path_table_input)    
table_img1 = upload_img_table(image_num = 1, path = path_table_input)  
table_img2 = upload_img_table(image_num = 2, path = path_table_input)
table_img3 = upload_img_table(image_num = 3, path = path_table_input)
table_img4 = upload_img_table(image_num = 4, path = path_table_input)
table_img5 = upload_img_table(image_num = 5, path = path_table_input)
table_img6 = upload_img_table(image_num = 6, path = path_table_input)
table_img = pd.concat([table_img0,table_img1,table_img2,table_img3,table_img4,table_img6],axis=0)
```

```python
import random
table_pheno = tabel_train_temp.loc[tabel_train_temp['pheno_class']==1,:]
table_wt = tabel_train_temp.loc[tabel_train_temp['pheno_class']==0,:]
random.shuffle(table_wt.index.values)
table_train_prop = pd.concat((table_pheno,table_wt.sort_index()[0:70]),0)
# duplicate table proprites
table_train_prop = table_train_prop.iloc[:,1:]
table_train_prop['label']  = table_train_prop.label.values - 1 
table_train_prop = table_train_prop.reset_index()
label_train = table_train_prop.label.values
image_number = table_train_prop.image_number.values
```

```python
def mereg_prop_granularity(table,table_imgg,index):
    table_init = table.iloc[:0,:].copy()
    table_img_init = table_imgg.iloc[:0,:].copy()
    tabel_con = pd.concat([table_init,table_img_init],axis=1)
    for i in range(len(index)):
        table_dup_temp = pd.concat([table.loc[(table['label']==index[i]) & 
                                              (table['image_number']==image_number[i]),:]]*80,ignore_index=True).reset_index()
        
        table_img_temp = table_imgg.loc[(table_imgg['image_group']==label_train[i]) & 
                                        (table_imgg['fov_number']==image_number[i]),:].reset_index()
        table_con_temp = pd.concat([table_dup_temp,table_img_temp],axis=1)
        table_con_temp = table_con_temp.iloc[:,2:]
        tabel_con = pd.concat([tabel_con,table_con_temp],axis=0)
    return tabel_con
table_sum_train = mereg_prop_granularity(table_train_prop,table_img,label_train)    
```
